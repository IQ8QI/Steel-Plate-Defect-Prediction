### Select targets and features

```python
targets = train_csv.columns[-7:]
features = train_csv.columns[1:-7]
train_features = train_csv[features]
train_targets = train_csv[targets]
test_features = test_csv[features]
train_targets
```

```python
train_features
```

### Do min-max normalization over data
Create norm_train_features from train_features.

Create norm_test_features from test_features.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_features)
norm_train_features = pd.DataFrame(scaler.transform(train_features), columns = train_features.columns)

scaler = MinMaxScaler()
scaler.fit(test_features)
norm_test_features = pd.DataFrame(scaler.transform(test_features), columns = test_features.columns)

norm_train_features
```

### Divide train_csv into train_data and test_data
This way we can do cross validation without spamin the competition

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(norm_train_features, train_targets, test_size=0.25, random_state=0)
train_data = pd.concat([X_train, y_train], axis=1)
test_data = pd.concat([X_test, y_test], axis=1)
train_data.head()
```

### Calculate volatility level of each explanatory variable
statistics is imported for mean and stdev

```python
import statistics
volatility_mat = pd.DataFrame(columns = features)
for column in volatility_mat.columns:
    average = statistics.mean(norm_train_features[column])
    standard_deviation = statistics.stdev(norm_train_features[column])
    volatility_value = abs(standard_deviation / average)
    volatility_mat.loc[0, column] = volatility_value

volatility_mat
```

### Find variables with volatility lower than 0.15
This variables will not be considered in further calculations.
Colmns of this variables are droped from the data.

```python
for column in volatility_mat.columns:
    if (vol := volatility_mat.loc[0, column]) < 0.15:
        print("Variable", column, "have volatility of", vol ,"and is lower than 0.15")
        # Dropping non-volatile variables from the data
        volatility_mat = volatility_mat.drop(column ,axis = "columns")
        
        X_train = X_train.drop(column ,axis = "columns")
        X_test = X_test.drop(column ,axis = "columns")
        
        train_csv = train_csv.drop(column ,axis = "columns")
        test_csv = test_csv.drop(column ,axis = "columns")
        
        norm_train_features = norm_train_features.drop(column ,axis = "columns")
        norm_test_features = norm_test_features.drop(column ,axis = "columns")
        
        train_features = train_features.drop(column ,axis = "columns")
        test_features = test_features.drop(column ,axis = "columns")

print("removed non-volatile variables")
```

### Calculate Person-r corelation matrix of the variables
scipy.stats is imported for pearsonr

```python
from scipy import stats
pearsonr_mat = pd.DataFrame(index = train_csv.columns[1:], columns = train_csv.columns[1:])
for col in pearsonr_mat.columns:
    for row in pearsonr_mat.index:
        if col == row:
            pearsonr_mat.loc[col, row] = 0
        else:
            result = stats.pearsonr(train_data[col], train_data[row])
            if result.statistic < result.pvalue:
                pearsonr_mat.loc[col, row] = 0
            else:
                pearsonr_mat.loc[col, row] = result.statistic

pearsonr_mat
```

### Find variables corelated to explanatory variables
Find corelations bigger than zero and sort them. First variable is the core variable, most corelated, according to its value other variables will be possibly removed in the future.

```python
explanatory_variables = {target : list() for target in targets}

for target in targets:
    for column in pearsonr_mat.sort_values(target, axis = "columns", ascending = False).columns:
        if pearsonr_mat.loc[target, column] != 0:
            explanatory_variables[target].append(column)

explanatory_variables
```

### Remove explanatory variables too related to each other
If explanatory variable is corelated more to explanatory variable most corelated to the explained variable, remove it.

```python
for target in targets:
    cor_df = pearsonr_mat.loc[explanatory_variables[target], explanatory_variables[target]]
    best_cor_value = pearsonr_mat.loc[target].max()
    for column in cor_df.columns:
        if cor_df.iloc[0].loc[column] > best_cor_value:
            explanatory_variables[target].remove(column)

print("number of features reduced")
explanatory_variables
```

### Calculate scores of RandomForestClassifier model for various max_leaf_nodes

```python
from sklearn.ensemble import RandomForestRegressor

max_leaf_nodes = [10, 25, 50, 75, 100, 150, 250]

scores = pd.DataFrame(columns = targets, index = [max_leaf_nodes])

for no_leafs in max_leaf_nodes:
    for target in targets:
        model = RandomForestRegressor(max_leaf_nodes = no_leafs, random_state = 1)
        model.fit(X_train[explanatory_variables[target]], y_train[target])
        scores.loc[no_leafs, target] = model.score(X_test[explanatory_variables[target]], y_test[target])

scores
```

```python
#scores.insert(-1, "avg", scores.mean(axis=1))

scores.mean(axis=1)
```

Best results are for 50 max_leaf_nodes

### Predict results with model made from full train data

```python
probabilities = pd.DataFrame(columns = [targets], index = test_csv.index)

for target in targets:
    model = RandomForestRegressor(max_leaf_nodes = 50, random_state = 1)
    model.fit(norm_train_features[explanatory_variables[target]], train_targets[target])
    probabilities[target] = model.predict(norm_test_features[explanatory_variables[target]])

probabilities
```
